worldbank dataset cleaning

```python
print("WORLD BANK: BASIC INFO")
print("Missing values (%): ")
missing_pct = (df_wb.isnull().sum() / len(df_wb) * 100).round(2)
print(missing_pct[missing_pct > 0].sort_values(ascending=False))
print()
print("Year range:", df_wb['year'].min(), "–", df_wb['year'].max())
print("Countries :", df_wb['country_name'].nunique())
print("Duplicates:", df_wb.duplicated().sum())
```

```python
#removing duplicates
before = len(df_wb)
df_wb = df_wb.drop_duplicates()
print(f"Duplicates removed: {before - len(df_wb)}")
```

```python
# Clean column names for df_wb
df_wb.columns = (
    df_wb.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_', regex=False)
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
    .str.replace('%', 'pct', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('[^a-z0-9_]', '', regex=True)
)
print("cleaned cloumns names")
for col in df_wb.columns:
    print(" ",col)
```

```python
# Remove years with excessive missing data and unnecessary years values
years_to_exclude = [2010, 2011, 2012, 2013, 2025]
df_wb = df_wb[~df_wb['year'].isin(years_to_exclude)]
```

```python
low_missing_cols  = ['gdp_current_usd', 'inflation_cpi_pct',
                     'gdp_growth_pct_annual', 'gdp_per_capita_current_usd',
                     'gross_national_income_usd', 'unemployment_rate_pct']

high_missing_cols = ['interest_rate_real_pct',
                     'government_expense_pct_of_gdp',
                     'government_revenue_pct_of_gdp',
                     'tax_revenue_pct_of_gdp',
                     'public_debt_pct_of_gdp']

# Interpolate low-missing columns within each country
for col in low_missing_cols:
    df_wb[col] = df_wb.groupby('country_name')[col].transform(
        lambda x: x.interpolate(method='linear').fillna(x.median())
    )
    print(f"  Interpolated: {col}")

# Flag high-missing columns — do NOT impute
for col in high_missing_cols:
    flag_col = col.replace('_pct_of_gdp','').replace('_pct','') + '_available'
    df_wb[flag_col] = df_wb[col].notna().astype(int)
    print(f"  Flagged     : {col} -> {flag_col}")

# Drop rows still missing GDP per capita or GDP growth (critical fields)
before = len(df_wb)
df_wb = df_wb.dropna(subset=['gdp_per_capita_current_usd', 'gdp_growth_pct_annual'])
print(f"\nRows dropped (critical fields missing): {before - len(df_wb)}")
print(f"Shape after: {df_wb.shape}")
```

```python
# Flag hyperinflation (>100%) — don't drop, these are real
df_wb['hyperinflation_flag'] = (df_wb['inflation_cpi_pct'] > 100).astype(int)

df_wb['inflation_capped'] = df_wb['inflation_cpi_pct'].clip(upper=100)

# Add crisis country label
crisis = ['Zimbabwe','Lebanon','Sudan','South Sudan','Venezuela, RB']
df_wb['is_crisis_economy'] = df_wb['country_name'].isin(crisis).astype(int)
```

unemployment data cleaning

```python
# Clean string columns for df_unemp
df_unemp['country_name'] = df_unemp['country_name'].str.strip().str.title()
df_unemp['sex'] = df_unemp['sex'].str.strip().str.title()
df_unemp['age_categories'] = df_unemp['age_categories'].str.strip().str.title()

print("String columns standardized")
print("  sex           :", sorted(df_unemp['sex'].unique()))
print("  age_categories:", sorted(df_unemp['age_categories'].unique()))
```

```python
# Standardize age_group values
age_map = {
  '15-24': '15-24 (Youth)',
  '25+': '25+ (Adults)',
  'Under 15': 'Under 15 (Children)'
}
df_unemp['age_group'] = df_unemp['age_group'].map(age_map)
print(df_unemp['age_group'].unique())
```

```python
# Drop 'indicator_name' — same value in every row
df_unemp = df_unemp.drop(columns=['indicator_name'])
```

```python
# Melt year columns into rows
year_cols = [c for c in df_unemp.columns if c.replace('.0','').isdigit()]

df_long = df_unemp.melt(
  id_vars=['country_name','sex','age_group','age_categories'],
  value_vars=year_cols,
  var_name='year',
  value_name='unemployment_rate'
)
df_long['year'] = df_long['year'].str.replace('.0','').astype(int)
print(f"   Shape: {df_long.shape}")
print(f"   Year range: {df_long['year'].min()} – {df_long['year'].max()}")
print(df_long.head(5))

# Country name harmonization - align with World Bank naming
country_map = {
  'Congo': 'Congo, Rep.',
  'Congo, Democratic Republic Of The': 'Congo, Dem. Rep.',
  'Egypt': 'Egypt, Arab Rep.',
  'Ivory Coast': "Cote d'Ivoire",
  'Korea, Republic Of': 'Korea, Rep.',
  "Korea, Democratic People'S Republic Of": 'Korea, Dem. Rep.',
  'Iran, Islamic Republic Of': 'Iran, Islamic Rep.',
  "Lao People'S Democratic Republic": 'Lao PDR',
  'Moldova, Republic Of': 'Moldova',
  'Hong Kong, China': 'Hong Kong SAR, China',
  'Macau, China': 'Macao SAR, China',
  'Palestinian Territories': 'West Bank and Gaza'
}
df_long['country_name'] = df_long['country_name'].replace(country_map)
print(f"\nCountry names harmonized")
```

```python
# Separate children data — ethically distinct from adult unemployment
df_children = df_long[df_long['age_categories'] == 'Children'].copy()
df_analysis = df_long[df_long['age_categories'] != 'Children'].copy()

# Document assumption:
# "Children (Under 15) unemployment data excluded from main analysis
# as ILO defines this differently from standard unemployment metrics.
# Retained separately for child labour discussion in ethics section."
```

Outlier Detection

```python
# IQR outlier detection for unemployment rate
Q1 = df_long['unemployment_rate'].quantile(0.25)
Q3 = df_long['unemployment_rate'].quantile(0.75)
IQR = Q3 - Q1

df_long['is_outlier'] = (
  (df_long['unemployment_rate'] < Q1 - 1.5*IQR) |
  (df_long['unemployment_rate'] > Q3 + 1.5*IQR)
).astype(int)

# DO NOT drop — flag only. Outliers = crisis events (e.g. COVID 2020)
print(f"IQR Outlier Detection:")
print(f"  Q1 (25th percentile): {Q1:.2f}%")
print(f"  Q3 (75th percentile): {Q3:.2f}%")
print(f"  IQR: {IQR:.2f}%")
print(f"  Lower bound: {Q1 - 1.5*IQR:.2f}%")
print(f"  Upper bound: {Q3 + 1.5*IQR:.2f}%")
print(f"  Outlier rows: {df_long['is_outlier'].sum()} ({df_long['is_outlier'].mean()*100:.1f}%)")
```

```python
# Identify and inspect extreme values (>80%)
suspect = df_long[df_long['unemployment_rate'] > 80]
print("Extreme unemployment rates (>80%):")
print(suspect[['country_name','year','sex','age_categories','unemployment_rate']])

# Flag rather than drop
df_long['extreme_rate_flag'] = (df_long['unemployment_rate'] > 80).astype(int)
print(f"\nExtreme rate rows: {df_long['extreme_rate_flag'].sum()}")
```

```python
total_missing = df_long['unemployment_rate'].isna().sum()
pct_missing   = round(total_missing / len(df_long) * 100, 2)
print(f"Missing unemployment_rate: {total_missing} rows ({pct_missing}%)")
print()

# Missing values are only in 2022/2023/2024 — interpolate within group
df_long = df_long.sort_values(['country_name','sex','age_categories','year'])
df_long['unemployment_rate'] = df_long.groupby(
    ['country_name', 'sex', 'age_categories']
)['unemployment_rate'].transform(
    lambda x: x.interpolate(method='linear').fillna(x.median())
)

remaining = df_long['unemployment_rate'].isna().sum()
print(f"After interpolation — remaining missing: {remaining}")
if remaining == 0:
    print("All unemployment rate gaps filled successfully!")

# NOW separate children data AFTER interpolation
df_children = df_long[df_long['age_categories'] == 'Children'].copy()
df_analysis = df_long[df_long['age_categories'] != 'Children'].copy()

print(f"\nSeparated datasets:")
print(f"  df_analysis: {df_analysis.shape} (Adults + Youth only)")
print(f"  df_children: {df_children.shape} (Children only)")
```

```python
print("UNEMPLOYMENT DATASET — CLEANING COMPLETE")
print(f"df_analysis shape : {df_analysis.shape}")
print(f"df_children shape : {df_children.shape}")
print(f"Year range        : {df_analysis['year'].min()} – {df_analysis['year'].max()}")
print(f"Countries         : {df_analysis['country_name'].nunique()}")
print(f"Sex categories    : {sorted(df_analysis['sex'].unique())}")
print(f"Age categories    : {sorted(df_analysis['age_categories'].unique())}")
print()
print("Missing values in df_analysis:")
print(df_analysis.isnull().sum())
```

```python
# Export cleaned datasets to CSV
print("Exporting cleaned datasets")

# World Bank cleaned data
df_wb.to_csv('../data/world_bank_cleaned.csv', index=False)
print(f"Saved: world_bank_cleaned.csv ({df_wb.shape[0]} rows, {df_wb.shape[1]} columns)")

# Unemployment long format (all data)
df_long.to_csv('../data/unemployment_long.csv', index=False)
print(f"Saved: unemployment_long.csv ({df_long.shape[0]} rows, {df_long.shape[1]} columns)")

# Unemployment analysis (adults + youth only)
df_analysis.to_csv('../data/unemployment_analysis.csv', index=False)
print(f"Saved: unemployment_analysis.csv ({df_analysis.shape[0]} rows, {df_analysis.shape[1]} columns)")

# Children unemployment (separate)
df_children.to_csv('../data/unemployment_children.csv', index=False)
print(f"Saved: unemployment_children.csv ({df_children.shape[0]} rows, {df_children.shape[1]} columns)")
```

```python
# Select key economic indicators
economic_cols = [
    'country_name', 'year',
    'gdp_per_capita_current_usd',
    'gdp_growth_pct_annual',
    'inflation_cpi_pct',
    'gdp_current_usd',
    'gross_national_income_usd',
    'unemployment_rate_pct',  # From World Bank (different from ILO data)
    'hyperinflation_flag',
    'is_crisis_economy'
]

df_merged = df_analysis.merge(
    df_wb[economic_cols],
    on=['country_name', 'year'],
    how='left',
    suffixes=('_ilo', '_wb')  # Distinguish unemployment sources
)

print("Merged shape:", df_merged.shape)
print("Countries:", df_merged['country_name'].nunique())
print("Missing values per column:")
print(df_merged.isnull().sum()[df_merged.isnull().sum() > 0])
```

Merge Statistics

```python
# Analyze merge quality
print("MERGE STATISTICS")
print("="*60)

# Countries in each dataset
unemp_countries = set(df_analysis['country_name'].unique())
wb_countries = set(df_wb['country_name'].unique())

matched = unemp_countries & wb_countries
unemp_only = unemp_countries - wb_countries
wb_only = wb_countries - unemp_countries

print(f"\nCountries in unemployment data: {len(unemp_countries)}")
print(f"Countries in World Bank data  : {len(wb_countries)}")
print(f"Matched countries             : {len(matched)} ({len(matched)/len(unemp_countries)*100:.1f}%)")
print(f"\nUnmatched (unemployment only) : {len(unemp_only)}")
if unemp_only:
    print(f"  {sorted(list(unemp_only))}")

# Rows with complete economic data
complete_rows = df_merged['gdp_per_capita_current_usd'].notna().sum()
print(f"\nRows with economic data       : {complete_rows}/{len(df_merged)} ({complete_rows/len(df_merged)*100:.1f}%)")
print(f"Rows missing economic data    : {len(df_merged) - complete_rows} ({(len(df_merged)-complete_rows)/len(df_merged)*100:.1f}%)")
```
